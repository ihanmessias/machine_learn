## Pr√©-processamento dos dados

- sklearn.model_selection.train_test_split:
> Usada para dividir um conjunto de dados em conjuntos de treinamento e teste para treinar e avaliar modelos de machine learning.

- sklearn.tree.DecisionTreeClassifier:
> √â um algoritmo de aprendizado de m√°quina baseado em √°rvore de decis√£o. O modelo de DecisionTreeClassifier cria uma √∫nica √°rvore de decis√£o para fazer previs√µes. Ele divide o conjunto de dados em subconjuntos com base em crit√©rios como entropia ou √≠ndice Gini, e faz previs√µes com base na maioria dos votos em cada folha da √°rvore.

- sklearn.linear_model.LogisticRegression:
> √â um algoritmo de aprendizado de m√°quina para classifica√ß√£o. O modelo de LogisticRegression usa uma fun√ß√£o log√≠stica para modelar a probabilidade de uma classe espec√≠fica. √â comumente usado para problemas de classifica√ß√£o bin√°ria, mas tamb√©m pode ser usado para problemas de classifica√ß√£o multiclasse atrav√©s de t√©cnicas como ‚Äúone-vs-rest‚Äù ou ‚Äúmultinomial‚Äù.

- sklearn.svm.SVC:
> √â um algoritmo de aprendizado de m√°quina baseado em vetores de suporte. O modelo SVC (Support Vector Classifier) tenta encontrar o hiperplano que melhor divide as classes no espa√ßo de caracter√≠sticas. Ele usa t√©cnicas como ‚Äúkernel trick‚Äù para lidar com dados n√£o linearmente separ√°veis.

- sklearn.ensemble.RandomForestClassifier:
> Algoritmo de aprendizado de m√°quina baseado em √°rvores de decis√£o. O modelo de RandomForest cria v√°rias √°rvores de decis√£o e as combina para obter previs√µes mais precisas ( Usa a m√©dia das previs√µes das √°rvores individuais para fazer previs√µes finais).

- sklearn.ensemble.VotingClassifier:
> √â um algoritmo de aprendizado de m√°quina que combina previs√µes de m√∫ltiplos modelos de aprendizado de m√°quina. O modelo de VotingClassifier faz previs√µes com base na maioria dos votos dos modelos individuais. Ele pode usar vota√ß√£o ‚Äúhard‚Äù, onde a classe com a maioria dos votos √© escolhida, ou vota√ß√£o ‚Äúsoft‚Äù, onde as probabilidades de classe s√£o somadas e a classe com a maior soma √© escolhida. √â √∫til para aumentar a robustez e a precis√£o do modelo ao combinar as for√ßas de v√°rios modelos diferentes.

- sklearn.metrics.classification_report:
> Relat√≥rio de m√©tricas de classifica√ß√£o, incluindo precis√£o, recall e F1-score para avaliar o desempenho de um modelo de classifica√ß√£o.

- sklearn.metrics.confusion_matrix:
> Matriz de confus√£o que mostra as contagens de verdadeiros positivos, verdadeiros negativos, falsos positivos e falsos negativos.

```python
# import machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # Model 1
from sklearn.linear_model import LogisticRegression # Model 2
from sklearn.svm import SVC # Model 3
from sklearn.ensemble import RandomForestClassifier # Model 4
from sklearn.ensemble import VotingClassifier

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# dataset split
X = df.drop(columns=['diagnosis']) # dataset independent variables (M,B)
y = df['diagnosis']  # Target data (what we want to predict)
# training and testing set (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

A escolha do n√∫mero 42 como valor para random_state ou qualquer outro n√∫mero inteiro n√£o tem um significado especial do ponto de vista da funcionalidade do algoritmo RandomForestClassifier. O importante √© que ele seja fixo para garantir que os resultados do modelo sejam reproduz√≠veis em diferentes execu√ß√µes do c√≥digo. Em programa√ß√£o e na comunidade de desenvolvedores, o n√∫mero 42 √© algumas vezes usado como uma ‚ÄúResposta para a Vida, o Universo e Tudo‚Äù em uma piada popular do livro ‚ÄúO Guia do Mochileiro das Gal√°xias‚Äù de Douglas Adams. No contexto da pergunta ‚Äúqual √© a resposta para a vida, o universo e tudo mais?‚Äù, a resposta foi calculada por um supercomputador como sendo o n√∫mero 42, embora o significado exato disso seja um mist√©rio, conforme a narrativa do livro.

## Constru√ß√£o dos Modelos

A raz√£o para usar essa metodologia √© que ela pode melhorar a precis√£o do modelo. Cada modelo individual pode ser bom em capturar certos padr√µes nos dados, mas pode perder outros. Ao combinar v√°rios modelos, voc√™ pode aproveitar os pontos fortes de cada modelo para obter um modelo final que √© mais preciso e robusto. Al√©m disso, isso pode ajudar a evitar o sobreajuste, pois os erros de um modelo podem ser compensados pelos acertos de outro modelo. Isso √© conhecido como ensemble learning.

```python
tree_clf = DecisionTreeClassifier()
lr_clf = LogisticRegression(max_iter=3000)
svm_clf = SVC()
rf_clf = RandomForestClassifier(random_state=42)

voting_clfs = VotingClassifier(estimators=[('tree', tree_clf), ('lr', lr_clf), ('svm', svm_clf), ('rf', rf_clf)], voting='hard')
voting_clfs.fit(X_train, y_train)
```

O par√¢metro max_iter=3000 define o n√∫mero m√°ximo de itera√ß√µes para o solucionador convergir. Se o modelo n√£o convergir em 3000 itera√ß√µes, ele ir√° parar e retornar um aviso. Aumentar este valor pode permitir que o modelo convirja para uma solu√ß√£o mais precisa, mas tamb√©m pode aumentar o tempo de treinamento.

Foi usado para corrigir o erro: "lbfgs failed to converge (status=1): STOP: TOTAL NO. of ITERATIONS REACHED LIMIT."

O erro ocorre quando o algoritmo de otimiza√ß√£o L-BFGS, usado na regress√£o log√≠stica, n√£o consegue encontrar a solu√ß√£o √≥tima dentro do n√∫mero m√°ximo de itera√ß√µes permitidas. Por padr√£o, o n√∫mero m√°ximo de itera√ß√µes √© definido como um valor relativamente baixo (por exemplo, 100). Se o algoritmo n√£o convergir para a solu√ß√£o √≥tima dentro desse limite, ele ir√° parar e retornar a mensagem de erro. Aumentar o n√∫mero m√°ximo de itera√ß√µes usando o par√¢metro max_iter=3000 d√° ao algoritmo mais tempo para encontrar a solu√ß√£o mais adequada.

## Avalia√ß√£o dos Modelos

Acur√°cia:
Ela representa a propor√ß√£o de previs√µes corretas feitas pelo modelo em rela√ß√£o ao total de previs√µes. Em outras palavras, a acur√°cia mede a precis√£o global do modelo.

A f√≥rmula para calcular a acur√°cia √©:



Classifica√ß√£o:

- Precision (Precis√£o): √â a porcentagem de previs√µes positivas corretas em rela√ß√£o ao total de previs√µes positivas feitas pelo modelo. Quanto mais alta, menos falsos positivos o modelo tem.
- Recall (Sensibilidade): √â a porcentagem de inst√¢ncias positivas reais que o modelo conseguiu capturar corretamente. Quanto mais alta, menos falsos negativos o modelo tem.
- F1-Score: √â uma m√©dia equilibrada entre precis√£o e recall. √â √∫til quando voc√™ quer uma m√©trica √∫nica que leve em considera√ß√£o tanto falsos positivos quanto falsos negativos.

```python
# classification_report and confusion matrix
for clf in (tree_clf, lr_clf, svm_clf, rf_clf, voting_clfs):
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    print(f'{clf.__class__.__name__}\n',classification_report(y_test, predictions))
    
    plt.figure(figsize=(6, 5))
    sns.heatmap(confusion_matrix(y_test, predictions),
        annot=True, fmt='d', cmap='Blues',
        xticklabels=['B', 'M'], yticklabels=['B', 'M'])
    plt.xlabel('Previsto')
    plt.ylabel('Real')
    plt.title(f'Matriz de Confus√£o {clf.__class__.__name__}:')
```

- Verdadeiros positivos (**B**, **B**): O modelo previu corretamente "n" inst√¢ncias como pertencentes √† classe B.
- Falsos positivos (**B**, **M**): O modelo previu erroneamente  "n" inst√¢ncias como pertencentes √† classe B, quando na verdade pertencem √† classe M.
- Falsos negativos (**M**, **B**): O modelo previu erroneamente "n" inst√¢ncias como pertencentes √† classe M, quando na verdade pertencem √† classe B.
- Verdadeiros negativos (**M**, **M**): O modelo previu corretamente "n" inst√¢ncias como pertencentes √† classe M.
